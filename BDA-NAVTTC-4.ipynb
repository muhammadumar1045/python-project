# --- Outliers Boundires Function ---

```python
def find_boundaries(df, variable, distance):
    Q1 =  df[variable].quantile(0.25)
    Q3 =  df[variable].quantile(0.75)
    IQR = Q3 - Q1
    lower_boundary = Q1 - (IQR * distance)
    upper_boundary = Q3 + (IQR * distance)
    return upper_boundary, lower_boundary
```

```python
sns.boxplot(dfS.RM)
```

```python
RM_upper_limit, RM_lower_limit = find_boundaries(dfS,'RM',1.5)
```

```python
RM_upper_limit, RM_lower_limit
```

# Let's create a Boolean vector to flag the outliers in RM:

```python
outliers_RM = np.where(dfS['RM'] > RM_upper_limit, True, 
                       np.where(dfS['RM'] < RM_lower_limit, True,False))
```

##  --- Best cluster ---

```python
 import matplotlib.pyplot as plt
```

```python
Sum_of_squared_distances = []
K = range(1,10)
for num_clusters in K :
    kmeans = KMeans(n_clusters=num_clusters)
    kmeans.fit(dfD)
    Sum_of_squared_distances.append(kmeans.inertia_)
    
plt.plot(K,Sum_of_squared_distances,'bx-')
plt.xlabel('Values of K') 
plt.ylabel('Sum of squared distances/Inertia') 
plt.title('Elbow Method For Optimal k')
plt.show()
```
